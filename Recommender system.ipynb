```python
#Filtrado basado en contenido
# Los dataFrames que tenemos son :
# candidatos_df = que son las canciones candidatas
# results_ds = ultimas 20 pistas escuchadas


matrix_results = results_ds.iloc[:, 1:].values
matrix_candidatos = candidatos_df.iloc[:, 1:].values

#Estandarizamos

scaler = StandardScaler()
last20_scaled = scaler.fit_transform(matrix_results)
candidatos_scaled = scaler.fit_transform(matrix_candidatos)

print(last20_scaled.mean(axis=0))
print(candidatos_scaled.std(axis=0))

```

```python
#Normalizar cada vector de caracteristicas (por filas)

#Magnitudes de cada vector (o pista)

last20_normalize = np.sqrt((last20_scaled*last20_scaled).sum(axis=1))
candidates_normalize = np.sqrt((candidatos_scaled*candidatos_scaled).sum(axis=1))

#Normalizacion

nLast20 = last20_scaled.shape[0]
nCandidates = candidatos_scaled.shape[0]
last20 = last20_scaled/last20_normalize.reshape(nLast20,1)
candidates = candidatos_scaled/candidates_normalize.reshape(nCandidates,1)

print(np.sqrt((last20*last20).sum(axis=1)))
print(np.sqrt((candidates*candidates).sum(axis=1)))

```

```python
#Calcular similitudes del coseno entre cada top-20 y cada una de las pistas candidatas

def cosine_similarity(X, Y):
    dot_product = np.dot(X, Y.T)  # Calcula el producto punto entre X e Y
    norm_X = np.linalg.norm(X, axis=1)  # Calcula la norma de X a lo largo del eje 1
    norm_Y = np.linalg.norm(Y, axis=1)  # Calcula la norma de Y a lo largo del eje 1
    norm_X = norm_X.reshape(-1, 1)  # Reformatea la norma de X para poder dividir de forma segura
    norm_Y = norm_Y.reshape(1, -1)  # Reformatea la norma de Y para poder dividir de forma segura
    cosine_sim = dot_product / (norm_X * norm_Y)  # Calcula la similitud del coseno
    return cosine_sim

cos_sim = cosine_similarity(last20,candidates)
cos_sim.shape
```

```python
#Obtener candidatos para una pista dada, comparando con cosine_similarity

def obtener_candidatos(position, cos_sim, num_cands, umbral = 0.8):
    #El umbral define si una pisat es apta o no
    
    index = np.where(cos_sim[position, :] >= umbral)[0]
    
    index = index[np.argsort(cos_sim[position, index])[::-1]]
    
    if len(index) >= num_cands:
        cands = index[0:num_cands]
    else:
        cands = index
    
    return cands
```

```python
for i in range(20):
    cands = obtener_candidatos(i , cos_sim, 5)
    print(f'{i} ==> pistas candidatas: {cands}, similitudes: {cos_sim[i, cands]}')
```
